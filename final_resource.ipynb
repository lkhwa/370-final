```python
import helper_functions as hf

df = hf.merge_id()
df
```

```python
top_100_words = hf.top_hundred(df[['text']])
```

```python
# Chosen words from the top 100 list
positive = ['love', 'friendly', 'free', 'fast', 'happy']
negative = ['wait', 'closed', 'busy', 'long', 'slow']
neutral = ['service', 'atmosphere', 'location', 'wifi', 'donuts']

word_list = positive + negative + neutral

tips_with_word = hf.get_tips(word_list, df)
tips_with_word
```

```python
top_100_business = hf.top_businesses_by_tip(df, word_list)
top_100_business
```

```python
# get top 100 most frequent words 
top_100_word_list = [x[0] for x in top_100_words]
top_100_word_list
```

```python
from os import path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

from wordcloud import WordCloud

d = path.dirname("__file__")

# read the mask image
coffee_mask = np.array(Image.open(path.join(d, "mask.png")))

# generate word cloud
wc = WordCloud(background_color="white", max_words=200, mask=coffee_mask)
wc.generate(str(top_100_word_list))

# store to file
wc.to_file(path.join(d, "coffee_wordcloud.png"))

# show
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
