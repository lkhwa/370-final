## Project Description
On Yelp, where anyone can submit and read reviews, businesses can and should leverage the user-submitted data to not only better understand their customer feedback, but the greater competitive market.Some parts of the customer feedback are easily quantified, such as the number of stars out of 5 given, the “useful”, “funny”, or “cool” votes. However, what the user includes in the text of their feedback is even more important, containing raw insights and even complaints about their experience. 

For this project, we wanted to focus on Yelp tips: shorter than reviews, but quick and straight to the point with suggestions. As characteristic of rating-oriented feedback, Yelp reviews are typically skewed towards the extreme ends of the 5-star spectrum, with users frequently leaving reviews only when they have an awesome or terrible experience. However, with tips not tied to a rating, we saw an opportunity for analyzing this more objective source of feedback.

**The purpose of our research project is to better understand what’s in a tip, and what its content may infer about a business’ overall rating**. We’ve scoped our project down to analyzing tips for **coffee shops**.

Our main research question is:
- Is there a relationship between the attributes mentioned by users in tips and the business’ overall rating?

Our sub-questions are:
- What are commonly mentioned business attributes in tips?
- Does the context of an attribute in a tip vary across different business ratings?

## Methods

From our top 100 words, we manually coded each word into **four categories**:
- **Positive**: conventionally positive words, such as adjectives (“friendly”), verbs (“love”), and features (“free”)
- **Negative**: conventionally negative words, such as adjectives (“slow”), verbs (“wait”)
- **Neutral**: business features or descriptors, not tied to a feeling, such as “location”
- **Eliminated**: stop-words (“go”, “get”), over-general words (“food”, “coffee”), or too niche words (“chicken”) 


```python
# Chosen words from the top 100 list
positive = ['love', 'friendly', 'free', 'fast', 'happy']
negative = ['wait', 'closed', 'busy', 'long', 'slow']
neutral = ['service', 'atmosphere', 'location', 'wifi', 'donuts']

word_list = positive + negative + neutral

tips_with_word = hf.get_tips(word_list, df)
tips_with_word
```
